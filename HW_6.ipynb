```python
# Load the profiles for the 1 M_sun star
profile_95 = pd.read_csv("data/1MsunData/profile.95.data")
profile_147 = pd.read_csv("data/1MsunData/profile.147.data")
profile_290 = pd.read_csv("data/1MsunData/profile.290.data")
profile_4903 = pd.read_csv("data/1MsunData/profile.4903.data")
profile_5632 = pd.read_csv("data/1MsunData/profile.5632.data")

profiles = [
    {
        "profile": profile_95,
        "model_number": 95,
        "description": "Ignition of H burning in core (ZAMS)",
    },
    {
        "profile": profile_147,
        "model_number": 147,
        "description": "Present-day Sun",
    },
    {
        "profile": profile_290,
        "model_number": 290,
        "description": "Exhaustion of H in core",
    },
    {
        "profile": profile_4903,
        "model_number": 4903,
        "description": "He core becomes non degenerate",
    },
    {
        "profile": profile_5632,
        "model_number": 5632,
        "description": "Envelope of star ejected",
    },
]
```

```python
plt.figure(figsize=(10, 8))
ax = sns.scatterplot(
    data=hist1, x="log_Teff", y="log_L", hue="model_number"
)
ax.xaxis.set_label_text("log Teff")
ax.yaxis.set_label_text("log L")
ax.invert_xaxis()

for profile_info in profiles[1:]:
    model_number = profile_info["model_number"]

    row = hist1[hist1["model_number"] == model_number]
    x = row["log_Teff"]
    y = row["log_L"]
    ax.scatter(x, y, marker="^", label=model_number)

    # Add text annotation
    ax.annotate(
        f"({model_number}) {profile_info['description']}",
        xy=(x, y),
        xytext=(10, -5),
        fontsize=12,
        textcoords="offset points",
    )
```

```python
sns.scatterplot(data=hist1, x="star_age", y="center_he4", color="blue", label="He4")
sns.scatterplot(data=hist1, x="star_age", y="center_h1", color="red", label="H1")
plt.gca().invert_xaxis()
```

```python
# Plot 25 Msun radius vs time plots
hist25_1 = hist25[hist25["model_number"] < 473]
hist25_2 = hist25[hist25["model_number"] > 472]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

sns.scatterplot(
    data=hist25_1, x="star_age", y="center_he4", color="blue", label="He4", ax=ax1
)
sns.scatterplot(
    data=hist25_2, x="star_age", y="center_h1", color="red", label="H1", ax=ax2
)
ax2.set_title("Parallel Plot")
# plt.gca().invert_xaxis()
```

```python
# Calculate t_end - t and take log
t_end = hist1["star_age"].max()
log_time_left = np.log10(t_end - hist1["star_age"])

# Create the plots with the new x-axis
sns.scatterplot(x=log_time_left, y=hist1["center_he4"], color="blue", label="He4")
sns.scatterplot(x=log_time_left, y=hist1["center_h1"], color="red", label="H1")

# Invert the x-axis to show evolution left-to-right
plt.gca().invert_xaxis()
```

```python
def plot_hist1(hist1, ax):
    sns.scatterplot(
        x=hist1["model_number"], y=hist1["center_he4"], color="blue", label="He4", ax=ax
    )
    sns.scatterplot(
        x=hist1["model_number"], y=hist1["center_h1"], color="red", label="H1", ax=ax
    )


fig, axes = plt.subplots(2, 2, figsize=(12, 10))

ax1 = axes[0, 0]
ax2 = axes[0, 1]
ax3 = axes[1, 0]
ax4 = axes[1, 1]

plot_hist1(hist1, ax1)
sns.scatterplot(x=hist1["star_age"], y=hist1["model_number"], ax=ax2)
plot_hist1(hist1[hist1["model_number"] < 400], ax3)
plot_hist1(hist1[hist1["model_number"] > 5000], ax4)
```

```python
hist25_1 = hist25[hist25["model_number"] < 473]
hist25_2 = hist25[hist25["model_number"] > 472]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

sns.scatterplot(
    data=hist25_1, x="star_age", y="center_he4", color="blue", label="He4", ax=ax1
)
sns.scatterplot(
    data=hist25_2, x="star_age", y="center_h1", color="red", label="H1", ax=ax2
)
ax2.set_title("Parallel Plot")
# plt.gca().invert_xaxis()
```

```python
step = 1000
total_plots = hist1["model_number"].max() // step
rows = int(total_plots // 2) + 1
cols = 2

fig, axes = plt.subplots(rows, cols, figsize=(15, 20))
axes = axes.flatten()


for i in range(0, int(hist1["model_number"].max()), step):
    sns.scatterplot(
        data=hist1[(hist1["model_number"] > i) & (hist1["model_number"] < i + step)],
        x="log_Teff",
        y="log_L",
        hue="model_number",
        ax=axes[i // step],
    )
    axes[i // step].invert_xaxis()
    axes[i // step].set_title(f"Model Number: {i} - {i + step}")

    # lebal 3 mid points

for profile_info in profiles:
    model_number = profile_info["model_number"]
    sns.scatterplot(
        data=hist1[hist1["model_number"] == model_number],
        x="log_Teff",
        y="log_L",
        marker="^",
        color="green",
        s=100,
        label=model_number,
        ax=axes[model_number // step],
    )

    # Add text annotation
    row = hist1[hist1["model_number"] == model_number]
    axes[model_number // step].annotate(
        f"({model_number}) {profile_info['description']}",
        xy=(row["log_Teff"].values[0], row["log_L"].values[0]),
        xytext=(10, -5),
        fontsize=14,
        textcoords="offset points",
    )

plt.tight_layout()
```

## Question 2

Starting from 290, the luminosity increases while the temperature decreases slightly because mathematically, $L = 4\pi R^2 \sigma T^4$, and a moderately drop in $T$ can be compensated by a sufficiently increase in $R$. 

```python
profile_352 = pd.read_csv("data/25MsunData/profile.352.data")
profile_474 = pd.read_csv("data/25MsunData/profile.474.data")
profile_704 = pd.read_csv("data/25MsunData/profile.704.data")
profile_808 = pd.read_csv("data/25MsunData/profile.808.data")
profile_1116 = pd.read_csv("data/25MsunData/profile.1116.data")

profiles_25 = [
    {
        "profile": profile_352,
        "model_number": 352,
        "description": "Exhaustion of H in core",
    },
    {
        "profile": profile_474,
        "model_number": 474,
        "description": "Core He ignition",
    },
    {
        "profile": profile_704,
        "model_number": 704,
        "description": "Core He ignition",
    },
    {
        "profile": profile_808,
        "model_number": 808,
        "description": "Core He ignition",
    },
    {
        "profile": profile_1116,
        "model_number": 1116,
        "description": "Core He ignition",
    },
]

step = 1000
total_plots = hist25["model_number"].max() // step
rows = int(total_plots // 2) + 1
cols = 2

fig, axes = plt.subplots(
    rows,
    cols,
    figsize=(
        15,
        rows * 6,
    ),
)
axes = axes.flatten()


for i in range(0, int(hist25["model_number"].max()), step):
    sns.scatterplot(
        data=hist25[(hist25["model_number"] > i) & (hist25["model_number"] < i + step)],
        x="log_Teff",
        y="log_L",
        hue="model_number",
        ax=axes[i // step],
    )
    axes[i // step].invert_xaxis()
    axes[i // step].set_title(f"Model Number: {i} - {i + step}")

    # lebal 3 mid points

for profile_info in profiles_25:
    model_number = profile_info["model_number"]
    sns.scatterplot(
        data=hist25[hist25["model_number"] == model_number],
        x="log_Teff",
        y="log_L",
        marker="^",
        color="green",
        s=100,
        label=model_number,
        ax=axes[model_number // step],
    )

    # Add text annotation
    row = hist25[hist25["model_number"] == model_number]
    axes[model_number // step].annotate(
        f"({model_number}) {profile_info['description']}",
        xy=(row["log_Teff"].values[0], row["log_L"].values[0]),
        xytext=(10, -5),
        fontsize=14,
        textcoords="offset points",
    )

plt.tight_layout()
```
